```python
import glob
import os
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from IPython.display import Image, display

# '/datatmp/Datasets/SunRGBD/SUNRGBD/xtion/sun3ddata/*/*/*/image/*.jpg'
dataset_dir = '/datatmp/Datasets/SunRGBD/SUNRGBD/xtion/sun3ddata'
rgb_images  = '*/*/*/image/*.jpg'
depth_images = '*/*/*/depth/*.png'

rgb_glob    = glob.glob( os.path.join(dataset_dir, rgb_images))
depth_glob  = glob.glob( os.path.join(dataset_dir, depth_images))

print('RGB images: ', (len(rgb_glob) ))
print('Depth images: ', (len(depth_glob) ))

print(type(Image( rgb_glob[1])))
idx = 546
display(Image( rgb_glob[idx] ), Image( depth_glob[idx] ))
```

```python
import shutil
import os

new_dir = '/datatmp/Datasets/SunRGBD_comp/'

train_imgs = range(0,2000)
test_imgs  = range (2000, 2300) 

idx = 0
for i in train_imgs:
    shutil.copy( rgb_glob[i]  ,os.path.join( new_dir, 'train', 'rgb' ,  str(idx)+'.png') )
    shutil.copy( depth_glob[i],os.path.join( new_dir, 'train', 'depth' ,str(idx)+'.png') )
    idx+=1

idx = 0
for i in test_imgs:
    shutil.copy( rgb_glob[i]  ,os.path.join( new_dir, 'test', 'rgb' ,  str(idx)+'.png') )
    shutil.copy( depth_glob[i],os.path.join( new_dir, 'test', 'depth' ,str(idx)+'.png') )
    idx+=1
```
