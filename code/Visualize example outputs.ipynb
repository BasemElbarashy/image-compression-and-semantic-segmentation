```python
import matplotlib.image as mpimg
import os
import matplotlib.pyplot as plt
import pickle

#imgs_dir = '/datatmp/Experiments/belbarashy/exps/seg_arch_for_compAndReconsrurction/lambda_2048.0/compressed_reconstructed_images/'
imgs_dir = '/datatmp/Experiments/belbarashy/exps/seg_comp_rgb_d_downby2/lambda_1.0/compressed_reconstructed_images/'
# 3 and 17
img_name = '3'

fig = plt.figure(figsize=(18, 15))

try:
    img = mpimg.imread(os.path.join(imgs_dir, img_name+'_orig.png'))
    plt.subplot(321)
    plt.title('Original')
    plt.imshow(img)
    plt.axis('off')
    fig.subplots_adjust(hspace=.05, wspace=.1)
except:
    pass
try:
    img = mpimg.imread(os.path.join(imgs_dir, img_name+'_depth.png'))
    plt.subplot(322)
    plt.title('Dense depth')
    plt.imshow(img)
    plt.axis('off')
    fig.subplots_adjust(hspace=.05, wspace=.1)
except:
    pass
try:

    img = mpimg.imread(os.path.join(imgs_dir, img_name+'_map_gt.png'))
    plt.subplot(323)
    plt.title('ground truth')
    plt.imshow(img)
    plt.axis('off')
    fig.subplots_adjust(hspace=.05, wspace=.1)
except:
    pass
try:
    img = mpimg.imread(os.path.join(imgs_dir, img_name+'_map_pred.png'))
    plt.subplot(324)
    plt.title('prediction')
    plt.imshow(img)
    plt.axis('off')
    fig.subplots_adjust(hspace=.05, wspace=.1)
except:
    pass
try:
    img = mpimg.imread(os.path.join(imgs_dir, img_name+'_reconstructed.png'))
    plt.subplot(325)
    plt.title('reconstruction')
    plt.imshow(img)
    plt.axis('off')
    fig.subplots_adjust(hspace=.05, wspace=.1)
except:
    pass
with open(os.path.join(imgs_dir, img_name+'_metrics.pkl'), "rb") as fp:
    metrics = pickle.load(fp)
print('========== \nImage metrics:')
print(metrics)
print('========== \nSize of compressed image: ')

try:
    print(os.stat(os.path.join(imgs_dir, img_name+'_compressed.bin')).st_size/1000, 'kB')
except:
    pass
```
